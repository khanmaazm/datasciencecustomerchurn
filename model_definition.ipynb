```
def get_model(model_algorithm, model_params):
  """Gets a model object from the sklearn library, with the specified parameters

    Parameters:
    model_algorithm: name of the model algorithm to use (e.g., ExtraTreesClassifier). Please not that for the function to be able to load the model, first it needs to be imported from the library.
    model_params: Dictionary to be passed to the model to use as parameters.

    Returns:
    Model object, with the specified parameters, ready for training and predicting.

   """
  assert model_algorithm in globals().keys(), 'Model type {} has not been imported'.format(model_algorithm)
  model = globals()[model_algorithm](**model_params)
  return model
```
